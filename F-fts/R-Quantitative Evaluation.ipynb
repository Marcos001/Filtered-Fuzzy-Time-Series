# Estimate the number of fuzzy sets
## Data: Alabama enrollment

```r
data = read.csv('../../data/csv/Enrollments.csv', sep=';')

# apply IMFS
imfs <- emd(data$Enrollments)
```

```r
choose_index(data$Enrollments, imfs$residue, file_save='Resultados/QE_amount_set_Alabama.csv')
```

## Data: TAIEX 1992

```r
# read
taiex = read.csv('../../data/csv/stock/TAIEX_1992.csv')

# apply IMFS
imfs <- emd(taiex$values)
```

```r
plot.ts(imfs$imf)
```

```r
IMFS <- imfs$imf[,4] + imfs$imf[,5] + imfs$residue
```

```r
choose_index(taiex$values, IMFS, file_save='Resultados/QE_amount_set_TAIEX.csv', limite=18)
```

## Data: DAX

```r
data = read.csv('../../data/csv/stock/GDAXI.csv')
head(data)

# apply IMFS
imfs <- emd(data$Close)

# visualize IMFs
plot.ts(imfs$imf)
```

```r
IMFS <- imfs$imf[,4] + imfs$imf[,5] + imfs$residue
```

```r
ini = Sys.time()
choose_index(data$Close, IMFS, limite=182, file_save='Resultados/QE_amount_set_DAX.csv')
fim = Sys.time()
cat('Execution Time:',(fim-ini),'\n')
```
